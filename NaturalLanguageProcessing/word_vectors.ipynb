```python
import pandas as pd
import numpy as np
import spacy
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
doc = nlp("subnautica")
vector = doc.vector

doc2 = nlp("where to find laser cutter subnautica")
vector2 = doc2.vector
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(vector, vector2)
```
